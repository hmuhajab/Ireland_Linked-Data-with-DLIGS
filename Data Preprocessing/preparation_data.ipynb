```python
# adding id to the json file to populate osi id (barony)


import pandas as pd
import json

# Load your CSV data into a DataFrame
csv_file_path = 'To add id from csv to the json for populating/baronies-_OSi_National_Statutory_Boundaries_-_2019.csv'  # Replace with your CSV file path
df = pd.read_csv(csv_file_path)

# Load your JSON data
json_file_path = 'To add id from csv to the json for populating/baronies.json'  # Replace with your JSON file path
with open(json_file_path, 'r') as f:
    json_data = json.load(f)

# Update the JSON data with the CO_ID
for entry in json_data:
    # Find the matching CO_ID for the county name
    matching_row = df[df['NAME_EN'].str.lower() == entry['name'].lower()]
    if not matching_row.empty:
        co_id = matching_row['LOGAINM_RE'].iloc[0]
        entry['LOGAINM_RE'] = int(co_id)

# Save the updated JSON data back to a file
updated_json_file_path = 'To add id from csv to the json for populating/updated_baronies.json'  # Replace with your desired path
with open(updated_json_file_path, 'w') as f:
    json.dump(json_data, f, indent=4)
```

```python
import pandas as pd
import json

# Load your CSV data into a DataFrame
csv_file_path = 'To add id from csv to the json for populating/townlands__OSi_National_Statutory_Boundaries_-_2019.csv' # Replace with your CSV file path
df = pd.read_csv(csv_file_path,  low_memory=False)

# Load your JSON data
json_file_path = 'To add id from csv to the json for populating/townlands1.json' # Replace with your JSON file path
with open(json_file_path, 'r') as f:
    json_data = json.load(f)

# Update the JSON data with the CO_ID
for entry in json_data:
    # Find the matching CO_ID for the county name
    matching_row = df[df['NAME_EN'].str.lower() == entry['name'].lower()]
    if not matching_row.empty:
        co_id = matching_row['LOGAINM_RE'].iloc[0]
        if pd.notna(co_id):  # Check if co_id is not NaN
            try:
                entry['LOGAINM_RE'] = int(co_id)  # Attempt to convert co_id to an integer
            except ValueError:  # Catch the ValueError if co_id is not an integer
                entry['LOGAINM_RE'] = -1  # Set a default value if co_id is not an integer
        else:
            entry['LOGAINM_RE'] = -1  # Set a default value if NaN
    else:
        entry['LOGAINM_RE'] = -1  # Set a default value if no match is found

# Save the updated JSON data back to a file
updated_json_file_path = 'To add id from csv to the json for populating/updated_townland.json'  # Replace with your desired path
with open(updated_json_file_path, 'w') as f:
    json.dump(json_data, f, indent=4)
    

    
    
```

```python
import json

# Load your JSON data
json_file_path = 'To add id from csv to the json for populating/updated_townland.json'  # Replace with your JSON file path
with open(json_file_path, 'r') as f:
    json_data = json.load(f)

# Update the JSON data by removing "LOGAINM_RE" key where its value is -1
for entry in json_data:
    if entry.get('LOGAINM_RE') == -1:
        del entry['LOGAINM_RE']  # This will remove the key if its value is -1

# Save the updated JSON data back to a file
updated_json_file_path = 'To add id from csv to the json for populating/updated_townland_2.json'  # Replace with your desired output JSON file path
with open(updated_json_file_path, 'w') as f:
    json.dump(json_data, f, indent=4)

print("JSON file updated. All 'LOGAINM_RE' entries set to -1 have been removed.")
```

```python
#to delete the space beteen names to avoid %20 when populating 

import json

# Function to replace spaces with underscores in name fields
def replace_spaces(data):
    if isinstance(data, dict):
        for key, value in data.items():
            if key == 'name' and isinstance(value, str):
                data[key] = value.replace(' ', '_')
            elif isinstance(value, dict) or isinstance(value, list):
                replace_spaces(value)
    elif isinstance(data, list):
        for item in data:
            replace_spaces(item)

# Read the original JSON file
with open('To add id from csv to the json for populating/updated_townland_2.json', 'r') as file:
    data = json.load(file)

# Modify the data
replace_spaces(data)

# Write the modified data to a new JSON file
with open('To add id from csv to the json for populating/updated_townland_2_cleaned.json', 'w') as file:
    json.dump(data, file, indent=4)

print("The data has been processed and the new JSON file is created.")
```
